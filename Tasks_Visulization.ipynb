```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('bikes.csv')
df
```

```python
sns.distplot(df['rented_bikes_count'],color='b',bins=20)
```

```python
sns.pairplot(df)
```

```python
sns.set(rc={'figure.figsize': [8, 12]}, font_scale=1.3)
sns.boxplot(x='season',y='rented_bikes_count',data=df)
```

```python
sns.violinplot(x='season',y='rented_bikes_count',data=df)
```

```python
sns.boxplot(x='season',y='rented_bikes_count',data=df,hue='weather')
```

```python
sns.barplot(x='season',y='rented_bikes_count',data=df)
```

```python

sns.stripplot(x='weather',y='rented_bikes_count',data=df,palette='viridis')
```

```python
sns.barplot(x='season',y='rented_bikes_count',data=df,estimator=np.median)
```

```python
sns.heatmap(df.corr(),annot=True)
```
